```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

%matplotlib inline

import warnings
warnings.filterwarnings("ignore")

RANDOM_SEED = 42
np.random.seed(RANDOM_SEED)
```

```python
df = pd.read_csv("Epileptic Seizure Recognition.csv")
df
```

```python
df.drop(columns=['Unnamed'], inplace=True)
df
```

```python
df['y'] = (df['y'] == 1).astype(int)
df 
```

```python
X = df.drop(columns=['y'])
y = df['y']
```

```python
def plots_of_cluster(X_tsne, X_umap, title=None, hue=None):
    plt.figure(figsize=(12, 4))

    plt.subplot(1, 2, 1)
    sns.scatterplot(x=X_tsne[:, 0], y=X_tsne[:, 1], hue=hue, palette='rainbow', s=15)
    plt.title('TSNE')

    plt.subplot(1, 2, 2)
    sns.scatterplot(x=X_umap[:, 0], y=X_umap[:, 1], hue=hue, palette='rainbow', s=15)
    plt.title('UMAP')

    plt.suptitle(title)

    plt.tight_layout()
    plt.show()

from sklearn.manifold import TSNE
import umap.umap_ as umap

tsne = TSNE(n_components=2, random_state=RANDOM_SEED)
X_tsne = tsne.fit_transform(X)

u = umap.UMAP(n_components=2, random_state=RANDOM_SEED)
X_umap = u.fit_transform(X)

plots_of_cluster(X_tsne, X_umap, title=f'Представление сниженной размерности', hue=y)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=RANDOM_SEED, stratify=y)
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
from sklearn.metrics import classification_report

def test_report(labels):
    print(classification_report(y_true=y_test, y_pred=labels))


from sklearn import metrics

def confusion_matrix(y_pred):
    confusion_matrix = metrics.confusion_matrix(y_true=y_test, y_pred=y_pred)

    cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = confusion_matrix, display_labels = [0, 1])

    cm_display.plot()
    plt.show() 

def report_and_matrix(y_pred):
    test_report(y_pred)
    confusion_matrix(y_pred)
```

Модели, которые будем использовать

```python
from sklearn.linear_model import LogisticRegression

def logistic_regression(X_train, X_test):
    y_pred = LogisticRegression(random_state=RANDOM_SEED, class_weight='balanced').fit(X_train, y_train).predict(X_test)
    report_and_matrix(y_pred)


from catboost import CatBoostClassifier 
from sklearn.utils.class_weight import compute_class_weight
 
classes = np.unique(y_train)
weights = compute_class_weight(class_weight='balanced', classes=classes, y=y_train)
class_weights = dict(zip(classes, weights))

def catboost(X_train, X_test):
    y_pred = CatBoostClassifier(logging_level='Silent', class_weights=class_weights).fit(X_train, y_train).predict(X_test)
    report_and_matrix(y_pred)


from tslearn.neighbors import KNeighborsTimeSeriesClassifier

def knn(X_train, X_test):
    y_pred = KNeighborsTimeSeriesClassifier(n_neighbors=5, weights='distance', metric='euclidean').fit(X_train, y_train).predict(X_test)
    report_and_matrix(y_pred)
```

## LogisticRegression

```python
logistic_regression(X_train, X_test)
```

## CatBoost

```python
catboost(X_train, X_test)
```

## KNeigbours

```python
knn(X_train, X_test)
```

## Используем быстрое преобразование Фурье

```python
def compute_fft(signal):
    fft_value = np.fft.fft(signal)
    fft_magnitude = np.abs(fft_value)[: len(signal) // 2]
    return fft_magnitude
```

```python
spectral_features_train = np.array([compute_fft(row) for row in X_train])
spectral_features_test  = np.array([compute_fft(row) for row in X_test])
```

## LogisticRegression FFT

```python
logistic_regression(spectral_features_train, spectral_features_test)
```

## CatBoost FFT

```python
catboost(spectral_features_train, spectral_features_test)
```

## KNeigbours FFT

```python
knn(spectral_features_train, spectral_features_test)
```

## Используем Вейвлет-преобразование

```python
import pywt

waveletname = 'bior1.3'

def compute_wt(signal):
    df_m_2D = pd.DataFrame(pywt.wavedec(signal, waveletname)).dropna(axis=1)
    df_m_2D_T = df_m_2D.T
    return  pd.melt(df_m_2D_T)['value'].T.values
```

```python
wt_features_train = np.array([compute_wt(row) for row in X_train])
wt_features_test  = np.array([compute_wt(row) for row in X_test])
```

## LogisticRegression WT

```python
logistic_regression(wt_features_train, wt_features_test)
```

## CatBoost WT

```python
catboost(wt_features_train, wt_features_test)
```

## KNeigbours FFT

```python
knn(wt_features_train, wt_features_test)
```

## Сгенерируем статистические фичи

```python
import tsfel

cfg = tsfel.get_features_by_domain('statistical')
```

```python
tsfel_features_train = np.array([tsfel.time_series_features_extractor(cfg, row, verbose=0) for row in X_train])
tsfel_features_test  = np.array([tsfel.time_series_features_extractor(cfg, row, verbose=0) for row in X_test])

tsfel_features_train = tsfel_features_train.reshape((tsfel_features_train.shape[0], tsfel_features_train.shape[2]))
tsfel_features_test  = tsfel_features_test.reshape((tsfel_features_test.shape[0], tsfel_features_test.shape[2]))
```

## LogisticRegression tsfel features

```python
logistic_regression(tsfel_features_train, tsfel_features_test)
```

## CatBoost tsfel features

```python
catboost(tsfel_features_train, tsfel_features_test)
```

## KNeigbours tsfel features

```python
knn(tsfel_features_train, tsfel_features_test)
```

## Заключение

* Видно, что boosting забивает все гвозди

* В задаче были использованы 4 вида выделения признаков, но можно их поделить на две группы:
    1) Признаки, не учитывающие временные изменения:  
        * fft - так как применяли не оконное преобразование - то получаем в спектре частоты, которые в принципе встречались в сигнале за всю его длительность;
        * tsfel фичи - статистические фичи;
    2) Учитывающие временные изменения:  
        * wt - вейвлеты учитывают как спектр, так и его изменения со временем;
        * исходные временные ряды - сами по себе отражают такие изменения;

Boosting лучше всех может учитывать признаки и временную зависимость в них из всех использованных алгоритмов, также за ним к этому стремится knn.  
Два этих алгоритма хорошо справлялись с исходными временными рядами, но при извлечении признаков другого вида начинали еще улучшать свои показатели.  

Но если для задачи достаточно использовать выделенные признаки, в которые не зашиты временные зависимости - можно просто применить к ним логистическую регрессию.
